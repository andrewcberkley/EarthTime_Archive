```python
cooked_data = {}
for row in raw_data:
    dst = row['Destination']
    year = row['Year']
    if dst in destinations:
        for o in origins:
            if o not in cooked_data:
                cooked_data[o] = {}
                for d in destinations:
                    cooked_data[o][d] = {'2010': 0.0, 
                                         '2011': 0.0, 
                                         '2012': 0.0, 
                                         '2013': 0.0, 
                                         '2014': 0.0, 
                                         '2015': 0.0,
                                         '2016': 0.0,
                                         '2017': 0.0}
            val = float(row[o].replace(",", "")) if row[o] != '..' else 0.0
            cooked_data[o][dst][year] = val
```

```python
def get_yty_totals(years):
    lrp = {}
    xp = [2010,2011,2012,2013,2014,2015,2016,2017]
    fp = []
    for y in xp:
        fp.append(years[str(y)])
    for i in range(2010,2018):
        lrp[str(i)] = np.interp(i,xp,fp)
    return lrp
```

```python
def get_delta_yty(years):
    delta = {}
    for i in range(2010,2017):
        delta[str(i+1)] = years[str(i+1)] - years[str(i)]
    return delta
```

```python
for o in origins:
    for d in destinations:
        cooked_data[o][d]['yty'] = get_yty_totals(cooked_data[o][d])
        cooked_data[o][d]['delta'] = get_delta_yty(cooked_data[o][d]['yty'])
```

```python
#Add +1 to the earliest and latest years to get this range
#(i.e. - if your data ranges from 2010-2017, the below range should be 2011-2018)

rows = {}
for i in range(2011,2018):
    rows[str(i)] = []
    for o in origins:
        for d in destinations:
            if o != d and cooked_data[o][d]['delta'][str(i)] > 0:
                row = [i, names_to_iso[o], names_to_iso[d], cooked_data[o][d]['delta'][str(i)]]
                rows[str(i)].append(row)
```

```python
def get_alpha_val(val):
    for i in range(len(alpha_bins) - 1):
        if val == alpha_bins[i]:
            return i/20.
        elif val > alpha_bins[i] and val <= alpha_bins[i+1]:
            return (i+1)/20.
    
def process_row(row):
    year, org, dst, val = row                
    org_wm = centroids_by_iso_alpha_3[row[1].upper()]['epsg_3857']
    dst_wm = centroids_by_iso_alpha_3[row[2].upper()]['epsg_3857']
    mid_wm = [((dst_wm[0] + org_wm[0]) / 2), ((dst_wm[1] + org_wm[1]) / 2)]  
    mid_offset = [0,0]                
    dist = math.sqrt(math.pow(dst_wm[0] - org_wm[0],2) +  math.pow(dst_wm[1] - org_wm[1],2))
    if abs(dst_wm[1] - org_wm[1]) > abs(dst_wm[0] - org_wm[0]):
        if dst_wm[1] > org_wm[1]:
            mid_offset[0] = 1
            mid_wm[0] += dist*0.5
        else:
            mid_offset[0] = -1
            mid_wm[0] -= dist*0.5
    else:
        if dst_wm[0] < org_wm[0]:
            mid_offset[1] = 1   
            mid_wm[1] += dist*0.5
        else: 
            mid_offset[1] = -1
            mid_wm[1] -= dist*0.5
    return {"year": year, 
            "org": org.lower(),
            "org_wm": org_wm,
            "dst_wm": dst_wm,
            "dst": dst.lower(),
            "mid_wm": mid_wm,
            "mid_offset": mid_offset,
            "export_val": math.ceil(val),
            'a_alpha': get_alpha_val(math.ceil(val))
            }
    
def process_results(results):
    collection = []
    for row in results:
        collection.append(process_row(row))                    
    return collection

def to_json(col, fname):
    with open(fname, "w") as f:
        json.dump(col, f)
```

```python
for year in range(2011,2018):
    fname = os.path.join(out_dir, str(year) + ".json")
    collection = process_results(rows[str(year)])
    to_json(collection, fname)
```
